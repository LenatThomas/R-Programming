```r
getwd()
IrisData = read.csv('iris.csv')
plot(
  IrisData$sepal_length ,
  IrisData$petal_length ,
  xlab = 'Sepal Length',
  ylab = 'Petal Length' ,
  main = 'Sepal Length vs Petal Length' ,
  pch = 20 ,
)

points(IrisData$sepal_length[IrisData$species == "setosa"], IrisData$petal_length[IrisData$species == "setosa"], col = "blue", pch = 20)
points(IrisData$sepal_length[IrisData$species == "versicolor"], IrisData$petal_length[IrisData$species == "versicolor"], col = "red", pch = 19)
points(IrisData$sepal_length[IrisData$species == "virginica"], IrisData$petal_length[IrisData$species == "virginica"], col = "green", pch = 18)

legend("bottomright", legend = unique(IrisData$species), col = c("blue", "red", "green"), pch = 18:20, title = "Species")
dev.off()
```

```r
MtCarsData = read.csv('mtcars.csv')
plot(MtCarsData$disp, MtCarsData$mpg, xlab = "Displacement", ylab = "Miles per Gallon",
     main = "Displacement vs Miles per Gallon")


points(MtCarsData$disp[MtCarsData$cyl == 4], MtCarsData$mpg[MtCarsData$cyl == 4], col = "red", pch = 18)
points(MtCarsData$disp[MtCarsData$cyl == 6], MtCarsData$mpg[MtCarsData$cyl == 6], col = "blue", pch = 19)
points(MtCarsData$disp[MtCarsData$cyl == 8], MtCarsData$mpg[MtCarsData$cyl == 8], col = "green", pch = 20)

abline(lm(mpg ~ disp, data = MtCarsData), col = "red")

legend("topright", legend = unique(MtCarsData$cyl), col = c("red", "blue", "green"), pch = 1:3, title = "Cylinders")
dev.off()
```

```r
getwd()
MtCarsData <- read.csv("mtcars.csv")
MtCarsData$cyl <- as.character(MtCarsData$cyl)
Table <- xtabs(~ am + cyl, data = MtCarsData)
barplot(Table, beside = TRUE, col = c("red", "blue"),
        legend.text = colnames(Table),
        xlab = "Number of Cylinders", ylab = "Count",
        main = "Cylinders vs. Transmission Type")


legend("topleft", legend = unique(MtCarsData$am), fill = c("red", "blue"), title = "Transmission Type")
dev.off()
```
