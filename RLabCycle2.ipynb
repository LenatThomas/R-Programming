```r
getwd()
IrisData = read.csv('iris.csv')
plot(
  IrisData$sepal_length ,
  IrisData$petal_length ,
  xlab = 'Sepal Length',
  ylab = 'Petal Length' ,
  main = 'Sepal Length vs Petal Length' ,
  pch = 20 ,
)

points(IrisData$sepal_length[IrisData$species == "setosa"], IrisData$petal_length[IrisData$species == "setosa"], col = "blue", pch = 20)
points(IrisData$sepal_length[IrisData$species == "versicolor"], IrisData$petal_length[IrisData$species == "versicolor"], col = "red", pch = 19)
points(IrisData$sepal_length[IrisData$species == "virginica"], IrisData$petal_length[IrisData$species == "virginica"], col = "green", pch = 18)

legend("bottomright", legend = unique(IrisData$species), col = c("blue", "red", "green"), pch = 18:20, title = "Species")
dev.off()
```

```r
MtCarsData = read.csv('mtcars.csv')
plot(MtCarsData$disp, MtCarsData$mpg, xlab = "Displacement", ylab = "Miles per Gallon",
     main = "Displacement vs Miles per Gallon")


points(MtCarsData$disp[MtCarsData$cyl == 4], MtCarsData$mpg[MtCarsData$cyl == 4], col = "red", pch = 18)
points(MtCarsData$disp[MtCarsData$cyl == 6], MtCarsData$mpg[MtCarsData$cyl == 6], col = "blue", pch = 19)
points(MtCarsData$disp[MtCarsData$cyl == 8], MtCarsData$mpg[MtCarsData$cyl == 8], col = "green", pch = 20)

abline(lm(mpg ~ disp, data = MtCarsData), col = "red")

legend("topright", legend = unique(MtCarsData$cyl), col = c("red", "blue", "green"), pch = 1:3, title = "Cylinders")
dev.off()
```

```r
getwd()
MtCarsData <- read.csv("mtcars.csv")
MtCarsData$cyl <- as.character(MtCarsData$cyl)
Table <- xtabs(~ am + cyl, data = MtCarsData)
barplot(Table, beside = TRUE, col = c("red", "blue"),
        legend.text = colnames(Table),
        xlab = "Number of Cylinders", ylab = "Count",
        main = "Cylinders vs. Transmission Type")


legend("topleft", legend = unique(MtCarsData$am), fill = c("red", "blue"), title = "Transmission Type")
dev.off()
```

```r
MtCarsData <- read.csv("mtcars.csv")
Table <- table(MtCarsData$cyl, MtCarsData$am)
barplot(Table, beside = TRUE, col = c("blue", "red"),
        main = "Cyclinders Vs Transmission Type",
        xlab = "Number of Cylinders", ylab = "Count",
        legend.text = c("Automatic", "Manual"))

legend("topright", legend = c("Automatic", "Manual"), fill = c("blue", "red"))
```

```r
Histogram <- hist(MtCarsData$mpg, breaks = 10, col = blues9, main = "Miles Per Gallon", xlab = "Miles Per Gallon", ylab = "Frequency")
mpgMeans <- mean(MtCarsData$mpg)
mpgStD <- sd(MtCarsData$mpg)
text(x = mpgMeans, y = max(Histogram$counts), labels = paste("Mean: ", round(mpgMeans, 2)), adj = c(0, -1))
text(x = mpgMeans, y = max(Histogram$counts), labels = paste("Standard Deviation: ", round(mpgStD, 2)), adj = c(0, -2))
```

```r
boxplot(hp ~ gear, data = MtCarsData, col = c("blue", "yellow", "red"), pch = c(1, 2, 3),
        main = "Horsepower (hp) by Number of Gears (gear)",
        xlab = "Number of Gears", ylab = "Horsepower")
legend("topright", legend = unique(MtCarsData$gear), fill = c("blue", "green", "red"), pch = c(1, 2, 3),
       title = "Gears")
outliers <- boxplot.stats(MtCarsData$hp)$out
text(1, outliers[1], labels = outliers[1], pos = 3)
text(2, outliers[2], labels = outliers[2], pos = 3)
text(3, outliers[3], labels = outliers[3], pos = 3)
```

```r

Covid <- read.csv("covid19.csv")
Covid$Date <- as.Date(Covid$Date)
Filtered <- Covid[Covid$Country.Region == "India", ]
TimeSeries <- ts(Filtered$Confirmed, start = min(Filtered$Date), end = max(Filtered$Date), frequency = 7)
DecomposedTimeSeries <- decompose(TimeSeries)
plot(TimeSeries, main = "COVID-19 Time Series for the India", ylab = "Confirmed Cases", xlab = "Date")
lines(DecomposedTimeSeries$trend, col = "blue", lwd = 2)
lines(DecomposedTimeSeries$seasonal, col = "red", lwd = 2)
legend("topright", legend = c("Observed", "Trend", "Seasonal"),
       col = c("black", "blue", "red"), lty = c(1, 1, 1), lwd = c(1, 2, 2))
print(summary(DecomposedTimeSeries))

```
